### Runge Kutta 4


El método RK (de orden s) tiene la siguiente expresión, en su forma más general:

$${\displaystyle y_{n+1}=y_{n}+h\,\sum _{i=1}^{s}b_{i}k_{i}} \qquad \qquad (1)$$


donde h es el paso por iteración, o lo que es lo mismo, el incremento ${\displaystyle \Delta t_{n}}$ entre los sucesivos puntos ${\displaystyle t_{n}}$ y ${\displaystyle t_{n+1}}$. Los coeficientes ${\displaystyle k_{i}}$ son términos de aproximación intermedios, evaluados en ƒ de manera local

$${\displaystyle k_{i}=f\left(t_{n}+h\,c_{i}\,,y_{n}+h\,\sum _{j=1}^{s}a_{ij}k_{j}\right)\quad i=1,...,s.} \qquad \qquad (2)$$

con ${\displaystyle a_{ij},b_{i},c_{i}}$ coeficientes propios del esquema numérico elegido, dependiente de la regla de cuadratura utilizada. Los esquemas Runge-Kutta pueden ser explícitos o implícitos dependiendo de las constantes ${\displaystyle a_{ij}}$ del esquema. Si esta matriz es triangular inferior con todos los elementos de la diagonal principal iguales a cero; es decir, ${\displaystyle a_{ij}=0}$ para ${\displaystyle j=i,...,s}$, los esquemas son explícitos.

De (1) y (2), para s=4, obtenemos el método de RK4:

\begin{equation}
\left\{
	       \begin{array}{ll}
		 x_{i+1} = x_i +\Delta x \\
		 y_{i+1} = y_i + \frac{\Delta x}{6} \left( K_1 + 2K_2 +2 K_3 +K_4 \right)
	       \end{array}
	     \right.
\end{equation}

donde: 

\begin{equation}
\left\{
	       \begin{array}{ll}
		 K_1 = f(x_i,y_i) \\
		 K_2 = f\left( x_i + \frac{\Delta x}{2},y_i + \frac{\Delta x}{2} K_1\right) \\
         K_3 = f\left( x_i + \frac{\Delta x}{2},y_i + \frac{\Delta x}{2} K_2\right) \\
         K_4 = f(x_i +\Delta x,y_i +\Delta x K_3)
	       \end{array}
	     \right.
\end{equation}

De aqui vemos que el método RK4 toma una mejor estimacion de la pendiente, ya que (dado un intervalo $[x_i,x_{i+1}]$) considera un promedio ponderado de la pendiente en el extremo izquierdo $K_1$, dos pendientes mejoradas en el punto medio $K_2$ y $K_3$ y una pendiente en el extremo derecho $K_4$. 

Considere el siguiente PVI:
\begin{equation}
\left\{
	       \begin{array}{ll}
		 y' = 2xy-x \equiv f(x,y)\\
		 y(0) = 0.4
	       \end{array}
	     \right.
\end{equation}

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rk4(f,x0,y0,xf,nd):
    x = np.zeros(nd)
    y = np.zeros(nd)
    dx = float(xf-x0)/nd
    x[0] = x0
    y[0] = y0

    for i in range(nd-1):
      x[i+1] = x[i] + dx
      k1 = f(x[i],y[i]) 
      k2 = f(x[i]+dx/2, y[i]+0.5*dx*k1)
      k3 = f(x[i]+dx/2, y[i]+0.5*dx*k2)
      k4 = f(x[i]+dx , y[i]+dx*k3)
      y[i+1]=y[i] + (dx/6)*(k1+2*k2+2*k3+k4)

    return x, y
```

```python
def f(x,y):
    return 2*x*y-x

def ex_sol(x):
    return (-1.0/10)*np.exp(pow(x,2))-0.5
```

```python
##Comparación
xx, yy = rk4(f,0,0.4,3.0,100) 
vsol = np.vectorize(ex_sol)
yex = vsol(xx)

plt.figure(figsize=(12,8))
plt.plot(xx,yy,'b*',label = 'rk4')
plt.plot(xx,yex,'g', label='real')
plt.grid(True)
plt.legend()
```

#### Lorenz Attractor
image.png

La ecuación diferencial de Lorenz se representa de la siguiente manera.

$${\frac{dx}{dt}=s(y-x)}$$
$${\frac{dy}{dt}=rx-y-xz}$$
$${\frac{dz}{dt}=xy-bz}$$

Donde:
$s=10,r=28, b=2.667$

Las condiciones iniciales son:
$x(0)=1, y(0)=1 , z(0)=1.05$

Para poder resolver el problema:
* Utilizar Diferencias Finitas para aproximar las derivadas. 
* Almacenar los valores en vectores.
* Utilizar el código de ayuda para plotear

```python
np.zeros(5)
```

```python
np.empty(5)
```

```python
def lorenz(x, y, z, s=10, r=28, b=2.667):
    x_dot = s*(y-x)
    y_dot = r*x-y-x*z
    z_dot = x*y-b*z

    return x_dot, y_dot, z_dot


dt = 0.01
num_steps = 10000

# valores iniciales
xs = np.empty(num_steps)
ys = np.empty(num_steps)
zs = np.empty(num_steps)

# adjuntar valores iniciales
xs[0], ys[0], zs[0] = (0. , 1. , 1.05)

#aproximación lineal
for i in range(num_steps-1):
  x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
  xs[i+1] = xs[i] + dt*x_dot
  ys[i+1] = ys[i] + dt*y_dot
  zs[i+1] = zs[i] + dt*z_dot
                               
# Plot
ax = plt.figure(figsize=(12,8)).add_subplot(projection='3d')

ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```

### Simple Linear Reg


<br>

**Ecución de la recta:**

$ \Large y = \beta_{0} + \beta_{1}x $

- $ \beta_{1} $ primero hay que calcular el coeficiente
- La pendiente de la recta

**B1 coef:**

$ \Large \beta_1 = \frac{\sum_{i=1}^{n}(x_i - \bar{x}) \times (y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2} $

- $ \beta_{0} $ coeficiente depende de la pendiente
- Representa la intersección Y: ubicación en la que la línea intercepta el eje Y

**B0 coef:**

$ \Large \beta_0 = \bar{y} - \beta_1 \times \bar{x} $

```python
import numpy as np
import matplotlib.pyplot as plt

class SimpleLinearRegression:

    def __init__(self):
        self.b0 = None
        self.b1 = None
    
    def fit(self, X, y):
        sup = np.sum((X-np.mean(X))*(y-np.mean(y)))
        inf = np.sum((X-np.mean(X))**2)
        self.b1 = sup/inf
        self.b0 = np.mean(y) - self.b1*np.mean(X)
        
    def predict(self, X):
        return self.b0 + self.b1*X
```

```python
X = np.arange(1,301)
y = np.random.normal(loc=X, scale=20)

plt.figure(figsize=(14,7))
plt.scatter(X, y, s=200, c='#087E8B', alpha=0.65)
plt.title('Source dataset', size=20)
plt.xlabel('X', size=14)
plt.ylabel('Y', size=14)
plt.show()
```

```python
plt.hist(y)
```

```python
def shuffle_split_data(X, y):
    arr_r = rd.rand(X.shape[0])
    split = arr_r < np.percentile(arr_r,80)

    X_train = X[split]
    y_train = y[split]
    X_test =  X[~split]
    y_test = y[~split]

    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = shuffle_split_data(X, y)
```

```python
model = SimpleLinearRegression()
model.fit(X_train, y_train)
preds = model.predict(X_test)
```

```python
def _mean_squared_error(y, y_hat):
    er = 0
    for i in range(len(y)):
      er = er + (y[i]-y_hat[i])**2

    return er/len(y)
```

```python
print(_mean_squared_error(y_test, preds))
```

```python
model_all = SimpleLinearRegression()
model_all.fit(X, y)
preds_all = model.predict(X)

plt.figure(figsize=(14,7))
plt.scatter(X, y, s=200, c='#087E8B', alpha=0.65, label='Source data')
plt.plot(X, preds_all, color='#000000', lw=3, label=f'Best fit line > B0 = {model_all.b0:.2f}, B1 = {model_all.b1:.2f}')
plt.title('Best fit line', size=20)
plt.xlabel('X', size=14)
plt.ylabel('Y', size=14)
plt.legend()
plt.show()
```
