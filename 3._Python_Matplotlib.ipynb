## Countour Plots

$$f(x,y)=sen(x)^{10}+cos(10+xy)\times cos(x)$$

```python
#función f
def f(x, y):
    return np.sin(x)**10 + np.cos(10+x*y)*np.cos(x)
```

```python
# nueva data
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X,Y=np.meshgrid(x,y)
Z=f(X,Y)
```

```python
# contour
plt.contour(X,Y,Z, colors='black')
```

```python
# contour + cmap
plt.contour(X,Y,Z, cmap='inferno')
plt.colorbar()
```
